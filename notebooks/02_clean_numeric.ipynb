## Ages

```python
df["ages"].unique()[:30]
```

#### replace 1/2 with .5

```python
ages_clean = df["ages"].astype(str).str.replace("½", ".5", regex=False)
# ages_clean.head()
```

#### Extract ranges 6-99

```python
range_extract = ages_clean.str.extract(r"(?P<age_min>\d+\.?\d*)-(?P<age_max>\d+\.?\d*)")
# range_extract.head()
```

#### Extract singles 6+

```python
single_extract = ages_clean.str.extract(r"(?P<age_min>\d+\.?\d*)\+?")
# single_extract.head()
```

#### code to change everything

```python
# create min age
df['age_min'] = pd.to_numeric(range_extract['age_min'].fillna(single_extract['age_min']), errors='coerce')

df['age_max'] = pd.to_numeric(range_extract['age_max'], errors='coerce')

df.loc[ages_clean.str.contains(r"\+$", regex=True), "age_max"] = 99

# df[["ages", "age_min", "age_max"]].sample(20)
```

# List Price

```python
df["list_price"] = pd.to_numeric(df["list_price"], errors="coerce")

df["list_price"] = df["list_price"].round(2)
```

# Num Reviews

```python
df["num_reviews"] = pd.to_numeric(df["num_reviews"], errors="coerce")

df["num_reviews"] = df["num_reviews"].fillna(0)

df["num_reviews"] = df["num_reviews"].astype(int)
```
